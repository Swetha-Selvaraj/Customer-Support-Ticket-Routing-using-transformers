```python
import gradio as gr
from transformers import pipeline

zero_shot_classifier = pipeline("zero-shot-classification")
```

```python

candidate_labels = ["technical issue", "billing inquiry", "product feedback"]

def classify_ticket(description):

    classification_result = zero_shot_classifier(description, candidate_labels)
    predicted_category = classification_result['labels'][0]


    with open('ticket_log.txt', 'a') as log_file:
        log_file.write(f"Description: {description}\n")
        log_file.write(f"Predicted Category: {predicted_category}\n")
        log_file.write("=" * 40 + "\n")

    return predicted_category


iface = gr.Interface(
    fn=classify_ticket,
    inputs="text",
    outputs="text",
    title="Support Ticket Classifier",
    description="Enter a description of the support ticket and get the predicted category.",
)

if __name__ == '__main__':
    iface.launch()
```
